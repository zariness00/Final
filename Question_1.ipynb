```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn import tree
```

```python
# load the SANK dataset:
sank_data=pd.read_csv("Sank.csv",header=0)
```

```python
print(sank_data.shape)
print(list(sank_data.columns))
```

```python
sank_data
```

```python
sank_data = sank_data.dropna()
```

```python
sank_data
```

```python
X = sank_data.loc[:, sank_data.columns != 'Alive']
y = sank_data.loc[:, sank_data.columns == 'Alive']
d = {'male': 1, 'female': 0}
X = X.replace({"Sex": d})

y = np.array(y)
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

## SVM Model 

```python
# SVM_Model = svm.SVC(kernel='linear')
SVM_Model = svm.SVC()
SVM_Model.fit(X_train, y_train)
```

```python
svm_parameters = {'kernel': ('linear', 'rbf'),'C':[1, 10, 20]}
SVM_GS_CV = GridSearchCV(SVM_Model, svm_parameters)
SVM_GS_CV.fit(X_train, y_train)
```

```python
#Best model score
print("Best model score SVM Model :",SVM_GS_CV.score(X_train, y_train))
#Best Parameters
print("Best Parameters SVM Model :",SVM_GS_CV.best_params_)
```

## Logistic Regression Model

```python
logreg = LogisticRegression(solver='liblinear',multi_class='auto')

# fit the model with data
logreg = logreg.fit(X_train,y_train)
```

```python
LogRM_GS_parameters = {'C':[1, 5, 10]}
LogRM_GS_CV = GridSearchCV(logreg, LogRM_GS_parameters)
LogRM_GS_CV.fit(X_train, y_train)
```

```python
#Best model score Logistic Regression 
print("Best model score Logistic Regression :", LogRM_GS_CV.score(X_train, y_train))
#Best Parameters Logistic Regression 
print("Best Parameters Logistic Regression :",LogRM_GS_CV.best_params_)
```

## Random Forest Model

```python
RandForC=RandomForestClassifier()
RandForC.fit(X_train,y_train)
```

```python
RandForC_GS_parameters = {'n_estimators':[1, 5, 10]}
RandForC_GS_CV = GridSearchCV(RandForC, RandForC_GS_parameters)
RandForC_GS_CV.fit(X_train, y_train)
```

```python
#Best model score Logistic Regression 
print("Best model score Random Forest :", RandForC_GS_CV.score(X_train, y_train))
#Best Parameters Logistic Regression 
print("Best Parameters Random Forest :",RandForC_GS_CV.best_params_)
```

# Naive_Bayes_Gaussian Model

```python
GuassianN_Model = GaussianNB()
GuassianN_Model.fit(X_train,y_train)
```

```python
GuassianN_GS_parameters = {}
GuassianN_GS_CV = GridSearchCV(GuassianN_Model, GuassianN_GS_parameters)
GuassianN_GS_CV.fit(X_train, y_train)
```

```python
#Best model score Logistic Regression 
print("Best model score Naive_Bayes_Gaussian Model :", GuassianN_GS_CV.score(X_train, y_train))
#Best Parameters Logistic Regression 
print("Best Parameters Naive_Bayes_Gaussian Model :",GuassianN_GS_CV.best_params_)
```

# Naive Bayes Multinomial Model

```python
MultinomialN_Model = MultinomialNB()
MultinomialN_Model.fit(X_train,y_train)
```

```python
MultinomialN_parameters = {}
MultinomialN_GS_CV = GridSearchCV(MultinomialN_Model, MultinomialN_parameters)
MultinomialN_GS_CV.fit(X_train, y_train)
```

```python
#Best model score Logistic Regression 
print("Best model score Naive Bayes Multinomial Model :", MultinomialN_GS_CV.score(X_train, y_train))
#Best Parameters Logistic Regression 
print("Best Parameters Naive Bayes Multinomial Model :",MultinomialN_GS_CV.best_params_)
```

# Decision Tree Model

```python
dtree_Model = tree.DecisionTreeClassifier()
dtree_Model.fit(X_train,y_train)
```

```python
dtree_parameters = {'criterion':['gini','entropy']}
dtree_GS_CV = GridSearchCV(dtree_Model, dtree_parameters)
dtree_GS_CV.fit(X_train, y_train)
```

```python
#Best model score Logistic Regression 
print("Best model score Decision Tree Model :", dtree_GS_CV.score(X_train, y_train))
#Best Parameters Logistic Regression 
print("Best Parameters Decision Tree Model :",dtree_GS_CV.best_params_)
```

### For the given DataSet Decision tree model suits best, because it's very fast compared to other models and with having score of around 99%. Random Forest algorithm seems to be performing very close to decision tree with best score of 97% whereas SVM seems to be performing worse with too much processing time and having best score of just around 78%.
